```python
import csv, sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('owid_covid_data.db')
c = conn.cursor()
```

```python
df = pd.read_csv('owid_covid_data.csv')
```

```python
df.to_sql('owid_covid_data', conn, if_exists='append', index = False)
```
